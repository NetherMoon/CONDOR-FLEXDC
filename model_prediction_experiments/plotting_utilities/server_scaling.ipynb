```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200
```

```python
server_size = [25,50,75,125,250,500]
sim_time = np.array([0.846,1.277,1.677,2.565,4.84,7.65])/1.944915
model_time = np.array([1,1,1,1,1,1])*5.5e-4
```

```python
plt.figure(figsize=(7,4.1))
plt.plot(server_size,sim_time,linestyle='dashed',c='Orange')
plt.scatter(server_size,sim_time,marker='*',c='Orange',label='Simulator')
plt.plot(server_size,model_time,linestyle='dashed',c='Blue')
plt.scatter(server_size,model_time,marker='*',c='Blue',label='CONDOR')
#plt.plot([25,500],[1,1],linestyle='dashed',c='Grey',label='Hour-Scale DR Requirement')
plt.plot([25,500],[1,1],linestyle='dashed',c='red',label='Hour-Scale DR Requirement')
plt.legend()
plt.ylabel('Bid Optimization Time [h]', fontsize=15)
plt.xlabel('Server Count', fontsize=15)
plt.title('Bid Optimization Scaling', fontsize=18)
plt.savefig('server_scaling.pdf',bbox_inches='tight')
```
