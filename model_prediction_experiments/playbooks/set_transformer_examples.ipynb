```python
from data_center_model import * 
import numpy as np
```

```python
# makes model
set_trans = SetTransformer(dim_input=6, num_outputs=1, dim_output=16)
```

```python
# makes three random "workloads" to test things
np.random.seed(42)
workload_1 = np.random.rand(1,10,6) # 1 batch, 10 workloads, of 6 variables long each 

output = set_trans(torch.tensor(workload_1).float())
print(output)
```

```python
# shows permutation invariance
np.random.seed(42)
workload_1 = np.random.rand(6) # new workload 1
workload_2 = np.random.rand(6) # new workload 2

workload_set_1 = np.zeros((1,10,6))
workload_set_1[0,0] = workload_1
workload_set_1[0,1] = workload_2

workload_set_2 = np.zeros((1,10,6))
workload_set_2[0,5] = workload_1
workload_set_2[0,2] = workload_2

output_1 = set_trans(torch.tensor(workload_set_1).float())
print(output_1)

output_2 = set_trans(torch.tensor(workload_set_2).float())
print(output_2)

print( (output_1-output_2)<0.000001) # floating points you know how they are 
```

```python
# shows zero-padding does not work with the S.T.
np.random.seed(42)
workload_1 = np.random.rand(6) # new workload 1
workload_2 = np.random.rand(6) # new workload 2

workload_set_1 = np.zeros((1,1,6))
workload_set_1[0,0] = workload_1
print(workload_set_1)


workload_set_2 = np.zeros((1,2,6))
workload_set_2[0,0] = workload_1
print(workload_set_2)


output_1 = set_trans(torch.tensor(workload_set_1).float())
print(output_1)

output_2 = set_trans(torch.tensor(workload_set_2).float())
print(output_2)

print( (output_1-output_2)<0.000001) # floating points you know how they are 
```
