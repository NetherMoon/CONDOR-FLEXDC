```python
# idk this will do stuff
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
from aqa_parsing_utilities import *
```

```python
sim_feature_directory = '../numpy_data'
stuff = np.load(sim_feature_directory +'/features_qos.npy')
```

```python
print(stuff[0,:])
```

```python
print(get_GD_file_names())
```

```python
new_data = []
print(stuff.shape)
sim_feature_index = 0
for mix in ['w3','w4','w5','w6','w7','w8']:
    for util in ['5','75','9']:
        cost,cost_power,cost_error,cost_qos,P_watts,R_watts,P_coef,R_coef = get_sim_hour_steps('../aqa_sim_data/realExperiment\\GDexp10-' + mix + '-util' + util + '.csv')
        for i in range(len(cost)):
            new_sample = [mix+'-'+ util,cost[i],cost_power[i],cost_error[i],cost_qos[i],P_watts[i],R_watts[i],P_coef[i],R_coef[i],stuff[sim_feature_index,0],stuff[sim_feature_index,1],stuff[sim_feature_index,2],stuff[sim_feature_index,3],stuff[sim_feature_index,4],stuff[sim_feature_index,5]]
            new_data.append(new_sample)
        sim_feature_index += 1
new_data = np.array(new_data)
```

```python
# now data is in form 
# # features are - Average Job Power, Average Job Time, Average Job # Nodes, Number of Jobs
dataset = pd.DataFrame(new_data)
header_titles = ['Sim Mix and Util',
                 'Cost',
                 'Cost Power',
                 'Cost Error',
                 'Cost QoS',
                 'P Watts',
                 'R Watts',
                 'P Coef',
                 'R Coef',
                 'Average Job Power',
                 'Average Job Time',
                 'Average Job # Nodes',
                 'Number of Jobs',
                 'Utilization',
                 'Average QOS']

dataset.to_csv('../numpy_data/new_model_data.csv',header=header_titles)
```
