```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
# https://www.thoughtco.com/using-pickle-to-save-objects-2813661
def read_gd_output_file_to_new_numpy_format(filename,workload_name,util):
    hour_sim_data = pd.read_csv(filename).to_numpy()
    p = hour_sim_data[:,8]
    r = hour_sim_data[:,9]
    cost = hour_sim_data[:,2]
    # block is num samples x 5 (workload name, util, p, r, cost)
    new_block = []
    for line in range(len(p)):
        new_block.append(np.array([workload_name,str(util),str(p[line]),str(r[line]),str(cost[line])]))
    new_block = np.array(new_block)
    print(new_block[0])
    return new_block
```

```python
# grabs data stuff or something
new_1 = read_gd_output_file_to_new_numpy_format('python_data/exp_11_1_data/GDexp11-gs-w7_util8(1).csv','W3',0.8)
new_2 = read_gd_output_file_to_new_numpy_format('python_data/exp_11_1_data/GDexp11-gs-w4_util8(1).csv','W4',0.8)
new_3 = read_gd_output_file_to_new_numpy_format('python_data/exp_11_1_data/GDexp11-gs-w5_util8(1).csv','W5',0.8)
new_4 = read_gd_output_file_to_new_numpy_format('python_data/exp_11_1_data/GDexp11-gs-w6_util8(1).csv','W6',0.8)
#new_5 = read_gd_output_file_to_new_numpy_format('python_data/exp_11_data/GDexp11-gs-w7_util8.csv','W7',0.8)

# old data
old = np.load('python_data/new_model_data_v2.npy')
print(old[0])
combined = np.concatenate((old,new_1,new_2,new_3,new_4),axis=0)
print(combined.shape)
np.save('python_data/new_model_data_v3.npy',combined)
```
